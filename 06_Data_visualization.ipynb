# Data visualization

- import the libraries

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
# import the titanic dataset from seaborn libraries

df = sns.load_dataset("titanic")
```

```python
df.plot(x = "age",y= "fare",kind="scatter")
plt.show()
```

```python
# import the color from color pricker
df.plot(x="age",y="fare",kind="scatter",color="#a32e9b")
```

```python
df.groupby("sex")["age"].mean()
```

```python
df.groupby("sex")["age"].mean().plot(kind='bar')
plt.ylabel("Average Fare")
plt.xlabel("Gender")
plt.show()
```

```python
# using seaborn library for plot

sns.barplot(y="survived",x="sex",data=df,hue="class")
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="class")
```

```python
sns.scatterplot(x="age",y="fare",data=df,hue="sex")
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
