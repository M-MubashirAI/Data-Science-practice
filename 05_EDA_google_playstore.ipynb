```python
import re

def convert_Size(Size):
    if isinstance(Size, str):
        if "k" in Size.lower():
            return float(re.sub("[kK]", "", Size)) * 1024
        elif "M" in Size:
            return float(Size.replace("M", "").strip()) * 1024 * 1024
        elif "Varies with device" in Size:
            return np.nan
    return Size
```

```python
# Befor apply the function column show the value
df["Size"]
```

```python
#Let's apply this funcion

df["Size"] = df["Size"].apply(convert_Size)
```

# install column
1. Remove sign +

```python
# Remove + sign
df["Installs"] = df["Installs"].apply(lambda x: x.replace("+", "")if "+" in str(x) else x)
```

```python
df["Installs"].value_counts()
```

```python
# Remove , sign
df["Installs"] = df["Installs"].apply(lambda x: x.replace(",", "") if "," in str(x) else x)
```

```python
# convert into integer

df["Installs"] = df["Installs"].apply(lambda x: int(x))
```

```python
df["Installs"].value_counts()
```

- Let's do binning in installs column

```python
# making a new column installs category Which will have the Installs category

bins = [-1,0,10,1000,10000,100000,1000000,100000000,10000000000]
labels = ["no","very low","low","moderate","more than moderate","high","very high","top notch"]

# # now Installs column convert into category
# df["Installs"] = df["Installs"].astype(int)
df["Installs_category"] = pd.cut(df["Installs"],bins=bins,labels=labels)
```
