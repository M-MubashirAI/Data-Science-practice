check that how many MBBs are there in one KB? and how to calculate?
1. convert KBS into M
2. then remove M from all numbers
3. Handle varies with device

```python
import pandas as pd

# Assuming the data is in a DataFrame called 'df'
# If it's not, you can create it from your data source
# df = pd.read_csv('your_data.csv')  # Uncomment this line if you need to read from a CSV file

# Function to convert size to MB
def convert_to_mb(size):
    if pd.isna(size) or size == 'Varies with device':
        return size
    elif isinstance(size, str):
        if size.endswith('M'):
            return float(size[:-1])
        elif size.endswith('k'):
            return float(size[:-1]) * 0.0009765625
    return size  # Return as is if format is unknown or already a number

# Apply the conversion function to the 'Size' column
df['Size_MB'] = df['Size'].apply(convert_to_mb)

# Separate numeric and non-numeric values
numeric_sizes = df['Size_MB'][df['Size_MB'].apply(lambda x: isinstance(x, (int, float)))]
non_numeric_sizes = df['Size_MB'][df['Size_MB'].apply(lambda x: not isinstance(x, (int, float)))]

# Count occurrences
numeric_counts = numeric_sizes.value_counts().sort_index()
non_numeric_counts = non_numeric_sizes.value_counts()

# Combine the results
size_counts = pd.concat([numeric_counts, non_numeric_counts])

# Print the results
print(size_counts)

# Optional: If you want to save the results to a CSV file
# size_counts.to_csv('app_sizes_mb.csv')
```

Remove the $ sign from all columns using pandas on above columns

```python
# Remove '$' from the 'Price' column
df['Price'] = df['Price'].str.replace('$', '', regex=False)

# Convert to numeric, coercing errors to NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Group by the cleaned price and count occurrences
price_counts = df['Price'].value_counts().sort_index()

# Print the results
print(price_counts)
```
