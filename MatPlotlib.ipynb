# **Bar plot Matplotlib**

```python
import matplotlib.pyplot as plt

y = [98,70,60,100,50,10,101]
x = ["Part1","Part2","Part3","Part4","Part5","Part6","Part7"]

color = ["red","green","blue","yellow","blue","orange","green"]

plt.bar(x ,y, color=color,edgecolor = "black")
plt.xlabel("Parts of Harry Potter",fontsize = 17)
plt.ylabel("Popularity",fontsize = 17)
plt.title("Popularity of difference Parts of Potter",fontsize = 20)
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
data = pd.read_excel("/content/expense3.xlsx")

# Check the first few rows of the DataFrame
print(data.head())

# Group the data by "Category" and sum the "Amount"
grouped = data.groupby("Category")["Amount"].sum()

# Convert the index to strings
grouped.index = grouped.index.astype(str)

# Print the grouped data
print(grouped)

# Plot the data
plt.plot(grouped.index, grouped.values)

# Set labels and title
plt.xlabel("Category")
plt.ylabel("Total Amount")
plt.title("Total Amount by Category")

# Show the plot
plt.show()
```

# **Pie Chart Plot**

# **Stack Plot Matplotlib**

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5,6,7]

week1 = [5,10,20,25,30,50,60,]
week2 = [10,20,30,25,40,60,50]
week3 = [8,20,25,50,60,70,50]
plt.stackplot(days,week1,week2,week3,labels=("week1","week2","week3"))
plt.legend()
plt.show()
```

# **A step Plot Matplotlib**

```python
x = ["day1","day2","day3","day4","day5","day6"]
y = [30,40,25,50,60,30]
plt.step(x,y,where="post",marker="o")
plt.show()
```

```python
data = pd.read_excel("/content/expense3.xlsx")
df = pd.DataFrame(data)
print(df)
group = df.groupby("Category").agg({"Amount":"sum"})
print(group)

plt.step(group.index ,group['Amount'], where="mid",marker="o")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
data = pd.read_excel("/content/expense3.xlsx")

# Create DataFrame
df = pd.DataFrame(data)
print(df)

# Group by Category and sum Amount
group = df.groupby("Category").agg({"Amount":"sum"})
print(group)
group.index= group.index.astype(str)

# Plotting
plt.step(group.index,group['Amount'], where="mid",marker="o")
plt.show()
```

# **Legend Matplotlib**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


a = [1,2,3,4,5,6]
y = [30,35,60,34,55,12]
y1 =[32,42,68,13,70,15]

plt.figure(figsize=[6,3])
plt.plot(a,y,label="male")
plt.plot(a,y1,label="female")
plt.legend(bbox_to_anchor = (0.8,1.2))
plt.show()
```

# **Sub Plot Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


x = [1,2,3,4,5,]
y = [45,36,50,29,55]
plt.subplot(1,2,1)#rows,columns,number of charts
plt.plot(x,y)
plt.title("Age")





d = [1,2,3,4,5,]
f = [60,65,50,59,70]
plt.subplot(1,2,2)
plt.bar(d,f)
plt.title("weight")


d = [1,2,3,4,5,]
f = [50,55,60,59,400]
plt.subplot(1,2,2)
plt.plot(d,f)

plt.title("hair color")
plt.suptitle("Employee data")
plt.show()
```
